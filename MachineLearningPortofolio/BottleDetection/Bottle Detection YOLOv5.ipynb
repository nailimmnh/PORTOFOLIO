```python
# Step 1: Install necessary libraries
!git clone https://github.com/ultralytics/yolov5  # Clone YOLOv5 repository
%cd yolov5
!pip install -qr requirements.txt  # Install YOLOv5 dependencies
!pip install opencv-python-headless  # Install OpenCV for video processing

# Import necessary libraries
import torch
import cv2
import numpy as np
from google.colab import files
from IPython.display import display, Image

# Step 2: Enable GPU Acceleration
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Step 3: Upload video file
# You will need to manually upload the video file from your laptop
uploaded = files.upload()  # This opens a file picker to select a video file

# Assuming only one video is uploaded, get the video file name
video_path = list(uploaded.keys())[0]
print(f"Uploaded video: {video_path}")

# Step 4: Load YOLOv5 model for object detection
# Load a pre-trained YOLOv5s model (the smallest version for faster inference)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
model.to(device)

# Step 5: Process video frame by frame for object detection
cap = cv2.VideoCapture(video_path)  # Initialize video capture

# Video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Output video writer
output = cv2.VideoWriter('output_video.avi', cv2.VideoWriter_fourcc(*'XVID'), fps, (frame_width, frame_height))

# YOLOv5 processes frames in real time and extracts bounding boxes and labels
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to RGB as YOLOv5 expects RGB format
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Run YOLOv5 model on frame
    results = model(img_rgb)

    # Extract bounding boxes, class labels, and confidences
    for det in results.pred[0]:
        # Unpack detection results
        x1, y1, x2, y2, conf, cls = det
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        # Draw bounding box and label on frame
        label = f"{results.names[int(cls)]} {conf:.2f}"
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write the processed frame to the output video
    output.write(frame)

# Release resources
cap.release()
output.release()

# Step 6: Download the output video
files.download('output_video.avi')
```
